```python
df_clean=df_clean.dropna(subset='anneedeplantation')
```

```python

# split the geo_point_2d column into two columns
df_clean[['lat','lon']] = df_removed['geo_point_2d'].str.split(',', expand=True)
df_clean.to_pickle("Tree_Grenoble_cleaned.pkl")
df_clean
```

```python
# division de la dataset
y_target=df_clean["anneedeplantation"]
X_feature=df_clean.drop('anneedeplantation', axis=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_feature,y_target,test_size=0.2,random_state=42)
```

```python
print(f'X_train={X_train.shape}\n X_test={X_test.shape}\n y_train={y_train.shape}\n y_test={y_test.shape}')
```

```python
#Median of yearsy_test
y_mediane=y_train.median()
y_mediane
```

```python
#créer le model
y_pred=np.zeros(y_test.shape)+y_mediane
f'y_pred={y_pred.shape}'
```

```python
#calculer les mae et mse
from sklearn.metrics import mean_absolute_error,mean_squared_error
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)

f'mae={mae}  mse={mse}'
#il faut trouver le modèle avec le plus petit possible de mae et mse
#pour ce la on va passer au pipline
```

```python
num_features=['adr_secteur','lat','lon']
ordinal_features=['stadededeveloppement','hauteurarbre']
categ_feature=['sous_categorie_desc','code_parent_desc','genre_bota','espece','collectivite','portarbre','typenature']
```

```python
from sklearn.pipeline import Pipeline,FeatureUnion
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OrdinalEncoder,OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
```

```python
num_pipline=Pipeline([('imputer',SimpleImputer(strategy='median')),
                      #impute les valeurs manquantes
                      ('scaler',StandardScaler())
                      #normalise les données
                      ])
                      
```

```python
#pipline pour les caracteriqtique ordinales
ord_pipline=Pipeline([('imputer',SimpleImputer(strategy='most_frequent')),
                      #imputer les valeurs manquantes
                      ('ordinal',OrdinalEncoder())
                      ])
```

```python
##pipline pour les caracteriqtique categor
cat_pipline=Pipeline([('imputer',SimpleImputer(strategy='most_frequent')),
                      #imputer les valeurs manquantes
                      ('onehot',OneHotEncoder(handle_unknown='ignore'))
                      #encoder les  données catg
                      ])
```

```python
cat_pipline
```

```python
num_pipline
```

```python
ord_pipline
```

```python
preprocessor=ColumnTransformer(transformers=[('num',num_pipline,num_features),
                                             ('ord',ord_pipline,ordinal_features),
                                             ('cat',cat_pipline,categ_feature)])
```

```python
preprocessor
```

```python
model_pipeline=Pipeline([('preprocessor',preprocessor)
                         ])
```

```python
model_pipeline
```

```python
model_pipeline.fit(X_train)
#X_train.columns
```

```python
x_train_pip=model_pipeline.fit_transform(X_train)
x_test_pip=model_pipeline.transform(X_test)
#(x_train_to array)
#stratify=X["genre"] #i faut ajouter sur les dex xtrain et y train
#fit : tu vas faire train determiner les parametres ..
#fit_transform 
#transform : appliquer la transformer

pd.DataFrame(x_train_pip.toarray())
```
