```python
num_features=['adr_secteur','lat','lon']
ordinal_features=['stadededeveloppement','hauteurarbre']
categ_feature=['sous_categorie_desc','code_parent_desc','genre_bota','espece','collectivite','portarbre','typenature']
```

```python
from sklearn.pipeline import Pipeline,FeatureUnion
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OrdinalEncoder,OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeRegressor,DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV # pour la division de dataset target et feature
from sklearn.metrics import mean_squared_error
```

```python
num_pipline=Pipeline([('imputer',SimpleImputer(strategy='median')),
                      #impute les valeurs manquantes
                      ('scaler',StandardScaler())
                      #normalise les données
                      ])
                      
```

```python
#pipline pour les caracteriqtique ordinales
ord_pipline=Pipeline([('imputer',SimpleImputer(strategy='most_frequent')),
                      #imputer les valeurs manquantes
                      ('ordinal',OrdinalEncoder())
                      ])
```

```python
##pipline pour les caracteriqtique categor
cat_pipline=Pipeline([('imputer',SimpleImputer(strategy='most_frequent')),
                      #imputer les valeurs manquantes
                      ('onehot',OneHotEncoder(handle_unknown='ignore'))
                      #encoder les  données catg
                      ])
```

```python
preprocessor=ColumnTransformer(transformers=[('num',num_pipline,num_features),
                                             ('ord',ord_pipline,ordinal_features),
                                             ('cat',cat_pipline,categ_feature)])
```

```python
preprocessor
```

```python
model_pipeline=Pipeline([('preprocessor',preprocessor),
                         
                         ])
```

GridSearchCV:

Grille de paramètres : Vous définissez une grille de paramètres à tester. Chaque combinaison possible de ces paramètres est évaluée.

Validation croisée : Pour chaque combinaison de paramètres, GridSearchCV utilise la validation croisée. Cela signifie que le modèle est formé et validé plusieurs fois en utilisant des sous-ensembles différents des données. Cette méthode permet de s'assurer que le modèle n'est pas seulement performant sur un sous-ensemble spécifique, mais qu'il généralise bien sur l'ensemble des données.

Sélection du meilleur modèle : Après avoir testé toutes les combinaisons, GridSearchCV sélectionne celle qui donne les meilleures performances en fonction d'une métrique spécifique (comme la précision, le rappel, etc.).

```python
#GridSearchCV

# Grille d'hyperparamètres
param_grid = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
}

```

```python
# Recherche par grille
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1, verbose=2)
grid_search.fit(x_train_pip, y_train)

# Récupération des résultats sous forme de DataFrame
results = pd.DataFrame(grid_search.cv_results_)
```

```python
# Affichage des résultats triés par score négatif de l'erreur quadratique moyenne
results = results[[ 'param_max_depth', 'param_min_samples_split', 'mean_test_score']]
results['mean_test_score'] = results['mean_test_score'].abs()  # Convertir en erreur quadratique moyenne positive
results = results.sort_values(by='mean_test_score',ascending=False)
print(results.head(10))  # Afficher les 10 meilleures configurations
```

```python
# Meilleure configuration trouvée
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
best_model.fit(x_train_pip,y_train)
```

```python
# Prédictions et évaluation
y_pred=best_model.predict(x_test_pip)
print("Meilleurs hyperparamètres :")
print(best_params)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error on Test Set: {mse}")
```

```python
from sklearn.svm import SVR
# Créer la pipeline avec le préprocesseur et un modèle SVM
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', SVR(kernel='linear', C=1))
])

# Définir le K-Fold cross-validator avec k=5
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Évaluer le pipeline avec cross_val_score
scores = cross_val_score(pipeline, X_feature, y_target, cv=kf, scoring='neg_mean_squared_error')

# Afficher les scores de chaque fold et la moyenne
print("Scores de chaque itération :", scores)
print("MSE moyenne :", scores.mean())
```

```python
from sklearn.ensemble import RandomForestRegressor
# Créer la pipeline avec le préprocesseur et un modèle RandomForestRegressor
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))
])

# Définir le K-Fold cross-validator avec k=5
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Évaluer le modèle avec cross_val_score et MSE
mse_scores = cross_val_score(pipeline, X_feature, y_target, cv=kf, scoring='neg_mean_squared_error')

# Afficher les scores de chaque itération et la moyenne
print("MSE de chaque itération (négatif) :", mse_scores)
print("MSE moyen (négatif) :", mse_scores.mean())
print("MSE moyen (positif) :", -mse_scores.mean())  # Inverser le signe pour obtenir le MSE positif

```
