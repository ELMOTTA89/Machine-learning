```python
df_clean=df_tree[['sous_categorie_desc','code_parent_desc','adr_secteur','genre_bota','espece','stadededeveloppement','anneedeplantation','collectivite','hauteurarbre','portarbre','typenature']]
df_removed=df_tree[['elem_point_id','nom','code','genre','categorie','sous_categorie','categorie_desc','code_parent','bien_reference','genre_desc',"variete","equipe",'remarques','raisondeplantation','traitementchenilles','courrier','identifiantplu','typeimplantationplu','intituleprotectionplu','anneeabattage','essouchement','diametrearbre','causeabattage','stationmetro','forme','structure','geo_point_2d']]

```

```python
#Remove missing value
nbr_missing_value=df_clean.isnull().sum()

print(nbr_missing_value)

print(df_clean.info())

df_clean.shape
df_clean.to_pickle("Tree_Grenoble_cleaned.pkl")
```

```python
df_clean=df_clean.dropna(subset='anneedeplantation')
```

```python

# split the geo_point_2d column into two columns
df_clean[['lat','lon']] = df_removed['geo_point_2d'].str.split(',', expand=True)
df_clean.to_pickle("Tree_Grenoble_cleaned.pkl")
df_clean
```

```python
# division de la dataset
y_target=df_clean["anneedeplantation"]
X_feature=df_clean.drop('anneedeplantation', axis=1)
X_feature['lat']=pd.to_numeric(X_feature['lat'])
X_feature['lon']=pd.to_numeric(X_feature['lon'])
X_feature['adr_secteur']=pd.to_numeric(X_feature['adr_secteur'])
#df[['col1', 'col2', 'col3']] = df[['col1', 'col2', 'col3']].apply(pd.to_numeric)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_feature,y_target,test_size=0.2,random_state=42)
```

```python
print(f'X_train={X_train.shape}\n X_test={X_test.shape}\n y_train={y_train.shape}\n y_test={y_test.shape}')
```

```python
#Median of yearsy_test
y_mediane=y_train.median()
y_mediane
```

```python
#créer le model
y_pred=np.zeros(y_test.shape)+y_mediane
f'y_pred={y_pred.shape}'
```

```python
#calculer les mae et mse
from sklearn.metrics import mean_absolute_error,mean_squared_error
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)

f'mae={mae}  mse={mse}'
#il faut trouver le modèle avec le plus petit possible de mae et mse
#pour ce la on va passer au pipline
```

```python
num_features=['adr_secteur','lat','lon']
ordinal_features=['stadededeveloppement','hauteurarbre']
categ_feature=['sous_categorie_desc','code_parent_desc','genre_bota','espece','collectivite','portarbre','typenature']
```

```python
from sklearn.pipeline import Pipeline,FeatureUnion
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OrdinalEncoder,OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeRegressor,DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV # pour la division de dataset target et feature
from sklearn.metrics import mean_squared_error
```

```python
num_pipline=Pipeline([('imputer',SimpleImputer(strategy='median')),
                      #impute les valeurs manquantes
                      ('scaler',StandardScaler())
                      #normalise les données
                      ])
                      
```

```python
#pipline pour les caracteriqtique ordinales
ord_pipline=Pipeline([('imputer',SimpleImputer(strategy='most_frequent')),
                      #imputer les valeurs manquantes
                      ('ordinal',OrdinalEncoder())
                      ])
```

```python
##pipline pour les caracteriqtique categor
cat_pipline=Pipeline([('imputer',SimpleImputer(strategy='most_frequent')),
                      #imputer les valeurs manquantes
                      ('onehot',OneHotEncoder(handle_unknown='ignore'))
                      #encoder les  données catg
                      ])
```

```python
cat_pipline
```

```python
num_pipline
```

```python
ord_pipline
```

```python
preprocessor=ColumnTransformer(transformers=[('num',num_pipline,num_features),
                                             ('ord',ord_pipline,ordinal_features),
                                             ('cat',cat_pipline,categ_feature)])
```

```python
preprocessor
```

```python
model_pipeline=Pipeline([('preprocessor',preprocessor),
                         
                         ])
```

```python
model_pipeline
```

```python
model_pipeline.fit(X_train)
#X_train.columns
```

```python
x_train_pip=model_pipeline.fit_transform(X_train)
x_test_pip=model_pipeline.transform(X_test)
#(x_train_to array)
#stratify=X["genre"] #i faut ajouter sur les dex xtrain et y train
#fit : tu vas faire train determiner les parametres ..
#fit_transform 
#transform : appliquer la transformer

pd.DataFrame(x_train_pip.toarray())
```

```python
pd.DataFrame(x_test_pip.toarray())
```

```python
model=DecisionTreeRegressor()
model.fit(x_train_pip,y_train)
```

```python
model_pipeline
```

```python
#prediction et evaluation
y_pred=model.predict(x_test_pip)
mse=mean_squared_error(y_test,y_pred)
print(f"mse={mse:.2f}")
mae=mean_absolute_error(y_test,y_pred)
print(f"mae={mae:.2f}")
print(np.sqrt(mse))

```

```python
#arbre de decision avec regression et après la classification (hyperparametre)
model_pipeline=Pipeline([('preprocessor',preprocessor),
                         ('regressor',DecisionTreeRegressor(random_state=42))
                         ])
model_pipeline
```

```python
#On a pas besion de FIT_TANSFORM OU TRANSFORM CAR ILS ONT ETE DEJÀ FAIT DANS LE PIPLINE 
model_pipeline.fit(X_train,y_train)
```

```python
#prediction et evaluation
y_pred=model_pipeline.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print(f"mse={mse:.2f}")
mae=mean_absolute_error(y_test,y_pred)
print(f"mae={mae:.2f}")
print(f"rmse={np.sqrt(mse)}")
```

```python
#GridSearchCV
# Grille d'hyperparamètres
param_grid = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
}

```

```python
# Recherche par grille
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1, verbose=2)
grid_search.fit(x_train_pip, y_train)

# Récupération des résultats sous forme de DataFrame
results = pd.DataFrame(grid_search.cv_results_)
```

```python
# Affichage des résultats triés par score négatif de l'erreur quadratique moyenne
results = results[[ 'param_max_depth', 'param_min_samples_split', 'mean_test_score']]
results['mean_test_score'] = results['mean_test_score'].abs()  # Convertir en erreur quadratique moyenne positive
results = results.sort_values(by='mean_test_score',ascending=False)
print(results.head(10))  # Afficher les 10 meilleures configurations
```

```python
# Meilleure configuration trouvée
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
best_model.fit(x_train_pip,y_train)
```

```python
# Prédictions et évaluation
y_pred=best_model.predict(x_test_pip)
print("Meilleurs hyperparamètres :")
print(best_params)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error on Test Set: {mse}")
```
